# **Introducción**

En la actualidad, el desempeño académico de los estudiantes es un tema central en el ámbito educativo y social, ya que se reconoce como un indicador clave del desarrollo personal y del progreso de las sociedades. Diversos factores influyen en el rendimiento académico, desde las características individuales de los estudiantes hasta las condiciones socioeconómicas, culturales y educativas que los rodean. Identificar y analizar estos factores permite no solo comprender las dinámicas del aprendizaje, sino también proponer estrategias efectivas para mejorar los resultados en distintos contextos.

Este estudio se enfoca en investigar la influencia de múltiples variables en el rendimiento académico de los estudiantes, incluyendo aspectos como la participación parental, el acceso a recursos, el nivel de motivación, la calidad del profesorado, y factores contextuales como el tipo de escuela, el género, y el ingreso familiar. Estas variables, aunque estudiadas de manera aislada en ocasiones, representan un conjunto complejo de interacciones que moldean las oportunidades de aprendizaje.

El objetivo de este trabajo es identificar patrones y relaciones significativas entre estas variables y el rendimiento académico, utilizando un enfoque cuantitativo y análisis de datos. Al explorar estos factores, se busca no solo entender las causas del éxito o fracaso escolar, sino también proponer recomendaciones para fortalecer los sistemas educativos y promover la equidad en el acceso a oportunidades educativas de calidad.

# **Marco teorico**

El marco teórico de esta investigación se construye en torno a diversas teorías y enfoques que abordan el rendimiento académico desde una perspectiva multidimensional. A continuación, se describen los conceptos clave y las investigaciones previas que sustentan este estudio.

**Factores familiares y participación parental**
La teoría del capital social de Coleman (1988) resalta la importancia de las relaciones familiares y su influencia en los resultados educativos. Según esta teoría, la participación activa de los padres en la educación de sus hijos, ya sea a través de la supervisión de tareas o la asistencia a eventos escolares, genera un entorno propicio para el aprendizaje. Estudios recientes corroboran que los estudiantes con mayor apoyo parental tienden a mostrar un rendimiento académico superior debido a la motivación y estructura que reciben en casa.

**Acceso a recursos educativos**
El acceso a recursos materiales, como libros, tecnología y espacios de estudio adecuados, es fundamental para el aprendizaje. Según la teoría del capital cultural de Bourdieu (1986), las familias con mayores recursos económicos tienen más probabilidades de proporcionar herramientas que favorecen el éxito académico de sus hijos. Este acceso no solo mejora el aprendizaje autónomo, sino que también reduce las desigualdades en los resultados educativos.

**Motivación intrínseca y extrínseca**
La teoría de la autodeterminación de Deci y Ryan (1985) distingue entre motivación intrínseca y extrínseca. La primera se relaciona con el deseo interno de aprender, mientras que la segunda se vincula a recompensas externas, como calificaciones o reconocimiento. Ambos tipos de motivación influyen en el rendimiento académico, siendo la motivación intrínseca un predictor más fuerte de logros a largo plazo.

**Calidad del profesorado y prácticas pedagógicas**
La teoría del aprendizaje social de Bandura (1977) enfatiza la importancia de los modelos en el aprendizaje. Los docentes, como figuras clave en la vida escolar, no solo transmiten conocimientos, sino que también influyen en las actitudes y comportamientos de los estudiantes. La calidad del profesorado, medida a través de su formación, experiencia y compromiso, se asocia consistentemente con mejores resultados académicos.

**Factores contextuales: género, tipo de escuela e ingreso familiar**
El contexto socioeconómico y cultural de los estudiantes tiene un impacto significativo en su rendimiento académico. Según Bronfenbrenner (1979) y su teoría ecológica del desarrollo humano, el aprendizaje ocurre dentro de un sistema de influencias interconectadas que incluyen la familia, la escuela y la comunidad. Variables como el género, el tipo de escuela (pública o privada) y el ingreso familiar afectan el acceso a oportunidades educativas, aunque estas diferencias pueden mitigarse mediante intervenciones dirigidas.

**Pruebas Estadísticas Utilizadas**

**Prueba ANOVA (Análisis de Varianza)**

La prueba ANOVA (Análisis de Varianza) se utiliza para comparar las medias de tres o más grupos independientes y determinar si existen diferencias significativas entre ellos. Es particularmente útil cuando se tiene más de dos grupos y se quiere verificar si la variable independiente afecta de manera distinta a los distintos grupos en cuanto a la variable dependiente.

La prueba ANOVA se utiliza cuando se quiere evaluar el efecto de una variable categórica sobre una variable continua.

**Prueba de Tukey**

La prueba de Tukey es una prueba post-hoc que se aplica después de realizar una ANOVA cuando se han encontrado diferencias significativas entre las medias de los grupos. Esta prueba permite identificar qué grupos específicos difieren entre sí. Es útil para realizar comparaciones múltiples entre los grupos sin aumentar el riesgo de cometer un error tipo I (falsos positivos).

La prueba de Tukey se utiliza cuando se ha encontrado una diferencia significativa en la ANOVA y se necesita identificar qué pares de grupos tienen diferencias significativas.

Interpretación:

Los resultados de Tukey proporcionan un valor p para cada comparación par. Si el valor p es menor que el nivel de significancia (usualmente 0.05), se considera que la diferencia entre los grupos comparados es significativa.

# **Metodologia**

El enfoque de esta investigación es cuantitativo, y se adopta un diseño correlacional para examinar las relaciones entre varias variables independientes (participación parental, acceso a recursos educativos, motivación académica, calidad del profesorado) y la variable dependiente (rendimiento académico). A través de un enfoque correlacional, el estudio busca identificar si existe una relación significativa entre las variables estudiadas y cómo estas pueden influir en el rendimiento académico de los estudiantes.

**1. Análisis de datos**

Prueba de Normalidad
Para determinar si las variables se ajustan a una distribución normal, se utilizará el test de Shapiro-Wilk. Si las variables siguen una distribución normal, se procederá con pruebas paramétricas; de lo contrario, se recurrirá a técnicas no paramétrica.

Análisis de Varianza (ANOVA)
Se aplica la prueba ANOVA de un solo factor para evaluar las diferencias en el rendimiento académico según los diferentes niveles de las variables independientes (participación parental, acceso a recursos, motivación y calidad del profesorado).

**2. Nivel de significancia**

Se utilizará un nivel de significancia de 0.05 (α = 0.05). Esto significa que si el valor p obtenido es menor que 0.05, se rechazará la hipótesis nula y se considerará que existen diferencias significativas entre los grupos.

**3. Interpretación de los resultados**

Prueba ANOVA:
Si el valor p es menor que 0.05, se concluirá que hay diferencias significativas entre las medias de los grupos en términos de rendimiento académico. De ser así, se procederá a realizar la prueba de Tukey para identificar qué grupos específicos presentan diferencias.

Prueba de Tukey:
Esta prueba permitirá identificar los pares de grupos con diferencias significativas en el rendimiento académico, proporcionándonos información sobre qué niveles de cada variable son más influyentes.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/diana462/Unidad-4/refs/heads/main/dataset.csv')
df
```

**Participacion parental**

```python
df['participacion_parental'].unique()
```

```python
# Si todos los grupos tienen una distribución normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

Interpretación:  La participación activa de los padres en el proceso educativo, como la supervisión de tareas y la asistencia a reuniones escolares, influye positivamente en el rendimiento académico de los estudiantes.

# **Resultados**

**Participación parental**

La participación activa de los padres tiene un impacto positivo en el rendimiento académico. Los estudiantes cuyos padres están más involucrados en actividades escolares y ayudan con tareas tienden a obtener mejores calificaciones. Este compromiso fomenta un ambiente de apoyo y motivación para los estudiantes, fortaleciendo su autoestima y habilidades de aprendizaje.

**Acceso a recursos**

Los estudiantes con acceso a recursos educativos, como libros, computadoras e internet, tienen un rendimiento significativamente superior. Estos recursos permiten el aprendizaje autónomo y refuerzan los conceptos aprendidos en clase, especialmente en áreas técnicas y de investigación. El acceso desigual a estos recursos sigue siendo un desafío importante en la equidad educativa.

**Motivación**

La motivación, especialmente la intrínseca, se identificó como un factor clave para el éxito académico. Los estudiantes motivados disfrutan el proceso de aprendizaje, se enfocan mejor en sus estudios y superan desafíos con una actitud positiva. Fomentar metas personales y el interés por las materias puede mejorar estos resultados.

**Actividades extracurriculares**

La participación en actividades extracurriculares, como deportes, música, arte o clubes escolares, está asociada con un mejor rendimiento académico y habilidades sociales. Estas actividades promueven la disciplina, el trabajo en equipo y el manejo del tiempo, proporcionando beneficios tanto académicos como personales.

**Horas de estudio semanales**

Se observó que un rango óptimo de 10 a 15 horas de estudio semanal está asociado con los mejores resultados académicos. Menos de 5 horas tienden a ser insuficientes, mientras que más de 20 horas no presentan mejoras significativas y pueden generar estrés académico.

**Nivel socioeconómico**

El nivel socioeconómico influye directamente en el rendimiento académico. Los estudiantes de familias con mayores ingresos o educación parental más alta suelen tener mejores resultados debido al acceso a recursos y apoyo adicional. Sin embargo, los programas de apoyo escolar han demostrado mitigar estas diferencias, mejorando el desempeño en grupos vulnerables.

**Estilo de enseñanza del profesor**

Los métodos de enseñanza activos, como aprendizaje basado en proyectos y discusiones interactivas, están asociados con mejores resultados académicos en comparación con métodos pasivos, como la memorización. Estos estilos de enseñanza involucran a los estudiantes, mejorando su pensamiento crítico y capacidad para resolver problemas.

**Asistencia escolar**

La asistencia regular es fundamental para un buen rendimiento académico. Los estudiantes con altas tasas de ausentismo presentan un desempeño inferior, ya que pierden clases clave y oportunidades de interacción. Un sistema de monitoreo y apoyo puede reducir el impacto del ausentismo.

**Entorno familiar**

Un entorno familiar estable, con comunicación abierta y apoyo emocional, fomenta un mejor desempeño académico. Los estudiantes que se sienten valorados en casa tienen una mayor autoconfianza y disposición para enfrentar retos escolares.

**Uso de tecnología**

El uso moderado de tecnología educativa (1-2 horas al día) tiene un impacto positivo en el rendimiento académico, especialmente en matemáticas y ciencias. Sin embargo, el uso excesivo para fines recreativos (>4 horas al día) puede reducir la productividad y los resultados académicos.
