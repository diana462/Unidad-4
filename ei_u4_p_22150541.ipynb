# **Introduccin**

Identificar y analizar estos factores permite no solo comprender las dinmicas del aprendizaje, sino tambin proponer estrategias efectivas para mejorar los resultados en distintos contextos.

Este estudio se enfoca en investigar la influencia de mltiples variables en el rendimiento acadmico de los estudiantes, incluyendo aspectos como la participacin parental, el acceso a recursos, el nivel de motivacin, la calidad del profesorado, y factores contextuales como el tipo de escuela, el gnero, y el ingreso familiar. Estas variables, aunque estudiadas de manera aislada en ocasiones, representan un conjunto complejo de interacciones que moldean las oportunidades de aprendizaje.

El objetivo de este trabajo es identificar patrones y relaciones significativas entre estas variables y el rendimiento acadmico, utilizando un enfoque cuantitativo y anlisis de datos. Al explorar estos factores, se busca no solo entender las causas del xito o fracaso escolar, sino tambin proponer recomendaciones para fortalecer los sistemas educativos y promover la equidad en el acceso a oportunidades educativas de calidad.

# **Marco teorico**

El marco terico de esta investigacin se construye en torno a diversas teoras y enfoques que abordan el rendimiento acadmico desde una perspectiva multidimensional. A continuacin, se describen los conceptos clave y las investigaciones previas que sustentan este estudio.

**Factores familiares y participacin parental**
La teora del capital social de Coleman (1988) resalta la importancia de las relaciones familiares y su influencia en los resultados educativos. Segn esta teora, la participacin activa de los padres en la educacin de sus hijos, ya sea a travs de la supervisin de tareas o la asistencia a eventos escolares, genera un entorno propicio para el aprendizaje. Estudios recientes corroboran que los estudiantes con mayor apoyo parental tienden a mostrar un rendimiento acadmico superior debido a la motivacin y estructura que reciben en casa.

**Acceso a recursos educativos**
El acceso a recursos materiales, como libros, tecnologa y espacios de estudio adecuados, es fundamental para el aprendizaje. Segn la teora del capital cultural de Bourdieu (1986), las familias con mayores recursos econmicos tienen ms probabilidades de proporcionar herramientas que favorecen el xito acadmico de sus hijos. Este acceso no solo mejora el aprendizaje autnomo, sino que tambin reduce las desigualdades en los resultados educativos.

**Motivacin intrnseca y extrnseca**
La teora de la autodeterminacin de Deci y Ryan (1985) distingue entre motivacin intrnseca y extrnseca. La primera se relaciona con el deseo interno de aprender, mientras que la segunda se vincula a recompensas externas, como calificaciones o reconocimiento. Ambos tipos de motivacin influyen en el rendimiento acadmico, siendo la motivacin intrnseca un predictor ms fuerte de logros a largo plazo.

**Calidad del profesorado y prcticas pedaggicas**
La teora del aprendizaje social de Bandura (1977) enfatiza la importancia de los modelos en el aprendizaje. Los docentes, como figuras clave en la vida escolar, no solo transmiten conocimientos, sino que tambin influyen en las actitudes y comportamientos de los estudiantes. La calidad del profesorado, medida a travs de su formacin, experiencia y compromiso, se asocia consistentemente con mejores resultados acadmicos.

**Factores contextuales: gnero, tipo de escuela e ingreso familiar**
El contexto socioeconmico y cultural de los estudiantes tiene un impacto significativo en su rendimiento acadmico. Segn Bronfenbrenner (1979) y su teora ecolgica del desarrollo humano, el aprendizaje ocurre dentro de un sistema de influencias interconectadas que incluyen la familia, la escuela y la comunidad. Variables como el gnero, el tipo de escuela (pblica o privada) y el ingreso familiar afectan el acceso a oportunidades educativas, aunque estas diferencias pueden mitigarse mediante intervenciones dirigidas.

**Pruebas Estadsticas Utilizadas**

**Prueba ANOVA (Anlisis de Varianza)**

La prueba ANOVA (Anlisis de Varianza) se utiliza para comparar las medias de tres o ms grupos independientes y determinar si existen diferencias significativas entre ellos. Es particularmente til cuando se tiene ms de dos grupos y se quiere verificar si la variable independiente afecta de manera distinta a los distintos grupos en cuanto a la variable dependiente.

La prueba ANOVA se utiliza cuando se quiere evaluar el efecto de una variable categrica sobre una variable continua.

**Prueba de Tukey**

La prueba de Tukey es una prueba post-hoc que se aplica despus de realizar una ANOVA cuando se han encontrado diferencias significativas entre las medias de los grupos. Esta prueba permite identificar qu grupos especficos difieren entre s. Es til para realizar comparaciones mltiples entre los grupos sin aumentar el riesgo de cometer un error tipo I (falsos positivos).

La prueba de Tukey se utiliza cuando se ha encontrado una diferencia significativa en la ANOVA y se necesita identificar qu pares de grupos tienen diferencias significativas.

Interpretacin:

Los resultados de Tukey proporcionan un valor p para cada comparacin par. Si el valor p es menor que el nivel de significancia (usualmente 0.05), se considera que la diferencia entre los grupos comparados es significativa.

# **Metodologia**

El enfoque de esta investigacin es cuantitativo, y se adopta un diseo correlacional para examinar las relaciones entre varias variables independientes (participacin parental, acceso a recursos educativos, motivacin acadmica, calidad del profesorado) y la variable dependiente (rendimiento acadmico). A travs de un enfoque correlacional, el estudio busca identificar si existe una relacin significativa entre las variables estudiadas y cmo estas pueden influir en el rendimiento acadmico de los estudiantes.

**1. Anlisis de datos**

Prueba de Normalidad
Para determinar si las variables se ajustan a una distribucin normal, se utilizar el test de Shapiro-Wilk. Si las variables siguen una distribucin normal, se proceder con pruebas paramtricas; de lo contrario, se recurrir a tcnicas no paramtrica.

Anlisis de Varianza (ANOVA)
Se aplica la prueba ANOVA de un solo factor para evaluar las diferencias en el rendimiento acadmico segn los diferentes niveles de las variables independientes (participacin parental, acceso a recursos, motivacin y calidad del profesorado).

**2. Nivel de significancia**

Se utilizar un nivel de significancia de 0.05 ( = 0.05). Esto significa que si el valor p obtenido es menor que 0.05, se rechazar la hiptesis nula y se considerar que existen diferencias significativas entre los grupos.

**3. Interpretacin de los resultados**

Prueba ANOVA:
Si el valor p es menor que 0.05, se concluir que hay diferencias significativas entre las medias de los grupos en trminos de rendimiento acadmico. De ser as, se proceder a realizar la prueba de Tukey para identificar qu grupos especficos presentan diferencias.

Prueba de Tukey:
Esta prueba permitir identificar los pares de grupos con diferencias significativas en el rendimiento acadmico, proporcionndonos informacin sobre qu niveles de cada variable son ms influyentes.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/diana462/Unidad-4/refs/heads/main/dataset.csv')
df
```

**Participacion parental**

```python
df['participacion_parental'].unique()
```

```python
# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

Interpretacin:  La participacin activa de los padres en el proceso educativo, como la supervisin de tareas y la asistencia a reuniones escolares, influye positivamente en el rendimiento acadmico de los estudiantes.
